Curiously, while the y-scale and time scale on each plot is wildly different, the *shapes* of the two plots are the same. This course, which is designed to teach you the *systems approach* to engineering modeling, should help you understand these two problems by seeing the *physical similarity* between them. This takes time, and it requires that we have a *physical understanding* of what is happening in each problem.

For the truck problem, we do have some tools we can use to understand the physics, so **understanding the physics behind the truck's coasting behavior is a good first step towards understanding the physics behind both problems.** 

The empirical model from Reading 6 is great, but it does little to illuminate the physics. Further, an empirical model doesn't tell us how (or whether) the truck would slow down differently if we added weight to its bed, or changed the size of its wheels. Physical or physics-based models are superior to purely empirical ones because they allow us to **predict the effects of design changes** on the behavior of our system. 

Put simply, a physics-based model is better because it *generalizes* to different model configurations better than an empirical one does.

While Newton's laws are a good starting place for developing a physics-based model of the truck, there are phenomena governing the truck's motion that we have not seen before. Your challenge today is to build the tools you will need to understand the physical phenomena that make the truck's behavior resemble the tank height behavior. We will look at the truck example together, and then you will be asked to develop a model of a new, similar system in today's assignment.

## Challenge: System Scoping

We will begin as we always do: by "scoping" a model for the truck system, showing the system's physical boundary along with notation indicating what crosses into and out of the system boundary, along with any relevant quantities that are stored. Here, we note that the truck is moving so it stores kinetic energy, and this energy leaves the system as the truck slows down in the form of losses due to friction.

## Challenge: Model Construction 

Treating the model of the truck this way, we might create an FBD of the truck, ,realizing that the only external force acting on it is friction.

Then, we might write Newton's second law as:

$$\sum \vec{F} = m\vec{\dot{v}} = N\hat{j} - mg\hat{j} - F_f \hat{i}$$

Recognizing that the  normal force and gravity will cancel one another, and that $\vec{v}=v\hat{i}$, we can write the x-component of Newton's second law to obtain:

$$m\dot{v} = -F_f$$

Our only model for friction so far is $F_f = \mu N = \mu mg$, and if we made that substitution and solved for the truck's accelertion, $\dot{v}$, we would get:

$$\dot{v} = -\mu g $$

This equation says that acceleration is a constant! This equation, unfortunately, **will *not* describe the behavior of the truck that we measured**, because we have *observed* that the acceleration of the truck is not a constant, and that its magnitude decreases with decreasing truck speed. However, we know there *must* be some sort of frictional force acting on the truck, or else it would keep rolling forever at a constant velocity, via Newton's 1st law.

To understand what is going on here, it is necessary to ask ourselves: "where is the friction coming from?"

The answer, as you might have guessed, is that the frictional force is a result of the interaction between the truck's axle and the truck's body. In order for the axle to turn, the body of the truck has a *bearing* installed for each. Bearings come in many shapes and sizes, but nearly all are *lubricated* interfaces between two parts that either slide or rotate relative to one another. Two types of bearing are shown below: the plain bearing and the ball bearing.

In our first pass at the model, it seemed that the friction slowing the truck must have come from the ground. There is, indeed, friction between the wheels and the ground, but it is difficult to know what it is, since the wheels are not sliding. They are *rolling without slip*, and the frictional force applied to the wheels by the ground is just enough at any moment to keep the wheels from sliding. Friction, as it turns out, is only well-modeled by a constant when two dry objects are sliding significantly relative to one another.

## Challenge: Model Scoping (again)

Because we have discovered that our original model scope was missing key information, we must revisit our process from the beginning, starting with scoping. The problem is not the overall system boundary, but the implication that we could treat the truck "monolithically" as one single element. We would do better to "pull out" key elements of our system until we expose the effects that define how our system behaves.

So, to unpack what is happening at the bearing interface between the truck's axles and its body, we will need to change our model scope. Specifically, we will need to treat three separate *elements* or subsystems in our model: the axles, the bearings, and the body of the truck. This is shown below:

Note that the only material difference between the two system scopes is that we are now explicitly differentiating between energy (if any) stored in the wheels, the energy (if any) stored in the bearings, and energy stored in the truck body. Graphically, we are now separating the truck into its relevant components, each of which will need a free-body diagram if we are to approach the problem using Newton's laws. 

Ultimately, we are **only** interested in the motion of the *body* of the truck. But in order to determine what its motion will be, we need to consider the motion of the body, the wheels/axles, and   the bearings separately. Breaking the scope apart into discrete elements helps us see what interactions may be important in our system. 

## Challenge: Model Construction 

We can begin our process of model construction with lists of key independent and dependent variables to help us keep track of what quantities we should see in our final equation. 


We'll start with a list of independent variables. Think of this list as keeping track of what you need to know about each element in your system scope in order to fully describe the system's behavior. Note that as we go through the process of model construction, **these lists may change** as we discover more independent variables we'll need to measure in order to produce our final model!

**Independent variables:**

* Truck body inertia/mass $m_c$
* wheel/axle inertia/mass $m_w$
* gravitational acceleration $g$
* frictional properties of bearings (not sure what these are yet, but we know they are important!!)

Next, we'll look at a list of relevant dependent variables in our model. This list will usually be short-- what are we trying to get an equation *for?*

**Dependent variables:**
* truck acceleration $a = \dot{v} = \ddot{x}$
* truck mass $v = \dot{x}$
* truck position $x$


Note that in the list above, $\dot{x}$ is used to denote $\frac{dx}{dt}$, and $\ddot{x}$ is used to denote $\frac{d^2x}{dt^2}$. The "dot" is a common notation for indicating time derivatives, and we will use it to keep equations that include derivatives easier to read.

Because we have numerical integration as a tool, we can focus our efforts on finding an equation for the dependent variable representing the *highest derivative* in our system. In this case, that's the acceleration of the truck body, $a = \dot{v} = \ddot{x}$. If we make it a goal to use Newton's laws to find an equation for $a$ in terms of only known quantities, we will be able to keep our eyes on the prize!

As we have done in the past, we will use the FBD of each element in our system as a guide for applying Newton's laws. Thankfully, the FBD for the truck does not change much with this new system scope.

The mass of the truck's body, now denoted $m_c$, used in the gravitational force term is different in this system scope than in our original formulation, since it only represents the mass of the truck's body. The x-component of the force from the bearing on the car body is denoted $F_{bcf}$ for the front bearing, and $F_{bcr}$ for the rear bearing. The forces supporting the car are now considered to be due to the bearing "holding up" the car body, and are denoted $N_{bcf}$ and $N_{bcr}$ respectively. First looking at Newton's second law for the truck body, since we observed no significant motion of the car in the $y$ direction, we can write the x-component as:

$$\sum F_x = m \dot{v} = -F_{bcf}-F_{bcr}$$

**THESE FORCES ARE BOTH UNKNOWN!!** However, with any luck, looking at the free-body diagrams for the axle and the bearing will allow us to unpack what they might be.

Let's continue by drawing an FBD for the axles. In particular, let's only draw the FBD for one of the axles, since they have the same *types* of forces acting on them, and their FBDs will look very similar. If we need to know the forces on the rear axle, and we feel they may be different than those on the front axle, we can perform that analysis later. 

The mass $m_w$ here represents the mass of each pair of wheels and the axle. It is here that we make the assumption, based on the identical appearance of the front and rear axles, that they have the same mass. Also notice that the axle's FBD includes the reaction force of the bearing on the axle, labeled $N_{bwf}$ on the FBD. There is also a normal force from the ground exerted on the axle to keep it from accelerating in the $y$ direction. This is labeled $N_{gwf}$. While these forces would be interesting to compute if we were worried about whether the axle was strong enough to support the car's weight, in our problem they aren't of any immediate relevance, since our  key dependent variable is the car's velocity $v$ in the x-direction.

Applying Newton's second law here, again in the x-direction, yields the scalar equation:

$$\sum F_x = F_{bwf}-F_{ff} = m\dot{v}$$

Where $F_{bwf}$ is the x-component of the force from the bearing on the axle. Notice that in this subsystem, the friction force and the "bearing force," which acts in the opposite direction as $F_{bcf}$ does on the car body, is what *drives the axle forward*. The frictional force is what keeps the point of contact between the wheel and the ground from sliding. It is this force that causes *rotation* of the wheel and axle. It does this by producing a *torque* or *moment* about the wheel's center of mass, or balance point.

## Torque or Moment


A torque (also called a 'moment') is applied to an object any time a force is applied at some distance from a reference point.

If the distance from the reference point to the application point of the force is defined by a position vector $\vec{r}$, as shown below:

Then the magnitude of the torque applied with respect to the reference point is the product of the position vector's magnitude and the component of the force vector that is *perpendicular* to the position vector:

$$T = r\cdot F_\perp$$

Think about the picture above and imagine that you squeezed the object shown with your thumb and forefinger at the reference point. If the force $\vec{F}$ was pointed directly perpendicular to the position vector $\vec{r}$, it would be hard to keep the object from rotating... and harder yet if the position vector got larger! But if the force vector was pointed either directly towards or directly away from your grip, the object would not "want" to rotate at all! This is the intuitive definition of the mathematical equation above, relating $F_\perp$ and $r$.

The definition of the component $F_\perp$ depends on the specific geometry of a particular problem, but if the angle $\theta$ is defined as in the figure above, then we can compute that $F_\perp=F\sin\theta$, where $F$ is the magnitude of the force vector $\vec{F}$. 

The *direction* of the torque is said to be perpendicular to the plane defined by the position and force vectors. Specifically, its direction can be surmised using the "right hand rule," in which one curls the fingers of the right hand from the position vector into the direction of the applied force. The direction of the torque is shown by the direction in which the thumb points. Note that this means that the applied *torque* will always point into or out of the page when considering planar motion.

This rule for computing the magnitude and direction of a *torque vector* $\vec{T}$ is a description of what is called the *vector cross product.* In terms of the vector cross product, torque is written:

$$\vec{T} = \vec{r}\times \vec{F}$$

Many times, the method of computing the direction and magnitude of torque explained above, as the "distance times the perpendicular component of force," is the easiest way. However, in cases where oblique forces are applied at odd distances from a reference point, computing the torque using the *component-wise* cross product can be more useful.

## The vector cross product

The vector cross product of two vectors $\vec{a}$ and $\vec{b}$ is shown as:

$$\vec{c} = \vec{a}\times\vec{b}$$

Any two vectors define a plane between them. In the cross product, the magnitude of $\vec{c}$ is given by the product of the magnitude $a$ with the component of $b$ which is *perpendicular* to the vector $\vec{a}$ in the plane defined between them. In the cross product, the vector $\vec{c}$ is always pointing into or out of the plane formed by $\vec{a}$ and $\vec{b}$.

The magnitude $c$ of the vector $\vec{c}$ is given by:

$$c = a b_\perp$$

And the direction of $\vec{c}$ is given by the *right hand rule*, in which the fingers of the hand are pointed in the direction of $\vec{a}$ and then curled into the direction of $\vec{b}$. After doing this, the direction of $\vec{c}$ is revealed by the direction of the thumb.

If the components of $\vec{a}$ and $\vec{b}$ are available, it is often easier to compute the magnitude and direction of $\vec{c}$ using the fact that the cross products of the unit vectors for the $x,y$ and $z$ directions for a standard, "right handed coordinate system," often denoted $\hat{i},\hat{j}$, and $\hat{k}$ respectively, are easy to compute.

Because $\hat{i},\hat{j},\hat{k}$ are all perpendicular to one another, and because they all have a magnitude of 1, their individual cross products are easy to compute using the right-hand rule. Get out your right hand and try each of these yourself!

\begin{align}
\hat{i}\times\hat{j} & = \hat{k} \\
\hat{j}\times\hat{k} & = \hat{i} \\
\hat{k}\times\hat{i} & = \hat{j} \\
\end{align}

Note that **the cross product is not commutative**, meaning that $\hat{j}\times\hat{i} \neq \hat{i}\times \hat{j}$. To prove this to ourselves, let's compute the reverses of the cross products above. Follow along!

\begin{align}
\hat{j}\times\hat{i} & = -\hat{k} \\
\hat{k}\times\hat{j} & = -\hat{i} \\
\hat{i}\times\hat{k} & = -\hat{j} \\
\end{align}

The right-hand rule also tells us that any two components in the same direction have a cross product of 0, e.g. $\hat{i}\times \hat{i}=0$. 

This may seem like a lot, but it makes computing cross products very quick and easy if components are available.


### Vector cross products using components: Example


Let's say we wished to compute the torque on a tree trunk "growing" out of the ground that is produced by a force $\vec{F}$ acting at its tip, with respect to the ground. The position vector from the ground to the application of the force is $\vec{r}$.

We know that $\vec{r} = 0.4\hat{i} + 1.2 \hat{j}$ meters, and that $\vec{F} = 100 \hat{i} - 200 \hat{j}$ Newtons. We could, of course, compute the magnitude of the torque using the formula $T = r\cdot F_\perp$, but we would need to compute the component of the force perpendicular to the position vector, which would require trigonometry. Instead, let's use the "standard" cross products for unit vectors above to speed things up. We can write:

$$\vec{T} = \vec{r}\times\vec{F} = (0.4\hat{i} + 1.2 \hat{j})\times(100 \hat{i} - 200 \hat{j})$$

Which, distributing the components using the "FOIL" (first, inside, outside, last) method, gives:

$$\vec{T} = (0.4\hat{i}\times 100 \hat{i}) + (0.4\hat{i} \times - 200 \hat{j}) + (1.2 \hat{j} \times 100 \hat{i}) + (1.2 \hat{j} \times - 200 \hat{j})$$

We can pull the magnitudes of each term out of the cross product to get:

$$\vec{T} = 0.4\cdot100(\hat{i}\times \hat{i}) + 0.4\cdot(- 200) (\hat{i} \times \hat{j}) + 1.2 \cdot 100 (\hat{j} \times  \hat{i}) + 1.2 \cdot (- 200) ( \hat{j} \times  \hat{j})$$

Then, using our unit-vector cross product relationships, we can compute the torque vector contribution of each term as:

$$\vec{T} = 0 -80\hat{k} -120\hat{k} + 0$$

Simplifying tells us that $\vec{T} = -200 \hat{k}N\cdot m$. 

Using the right-hand rule, we can confirm that indeed the torque should be into the page, or in the $-\hat{k}$ direction.

Now that we have an understanding of Torque and of the vector cross product, it is still unclear **how torque helps us understand what the bearing does to the car and the axle**. Fortunately, Torque can be analyzed using Newton's laws in much the same way force can.

## Newton's Second Law applied to angular momentum for a rigid body rotating about its center of mass in a plane


Newton's second law, a statement about the conservation of momentum, applies to *angular* as well as linear momentum. For a rigid body moving in a plane, angular momentum is defined as:

$$\vec{H} = J \omega \hat{k}$$

Where $\omega$ is the magnitude of the rigid body's *angular velocity*, and $J$ is the *mass moment of inertia* of the rigid body. Think of the mass moment of inertia like a "mass for rotation." The larger this number is, the harder it is to accelerate an object in rotation.

Mass moment of inertia is a scalar quantity that is related to how mass is distributed in an object.  A list of mass moments of inertia is available [here](https://en.wikipedia.org/wiki/List_of_moments_of_inertia), and the formal definition is given as:

$$J = \int r^2dm$$

Where "dm" represents an infinitesimal piece of mass that makes up the rigid body. It is rare that one needs to calculate a mass moment of inertia using this formula-- tables of mass moments of inertia for most common shapes are widely available, and most modern computer-aided design software can compute mass moments of inertia for parts you design. However, looking at the equation is helpful for understanding what moment of inertia *really is*. Imagine an object with all of its mass concentrated at a point. We could calculate that mass's *moment of inertia* with respect to some arbitrary axis as $J = mr^2$. By contrast, a wheel or a cube would have a different moment of inertia, since the mass of those objects is distributed in space.

Angular velocity, which has units of $\frac{rad}{s}$ in the SI system, has a *direction* either into or out of the plane of motion, according to the right-hand rule. with a standard coordinate system defined on a page as $x$ right and $y$ up, counterclockwise rotation is positive (out of the page), and clockwise rotation is negative (into the page). Newton's second law deals with the derivative of momentum. For a rotational system in a non-accelerating (also called "Newtonian") reference frame, we can write:

$$\sum \vec{T} =\frac{d\vec{H}}{dt}$$

Where **The angular momentum is computed about a rigid body's center of mass**. If moments (torques) are to be summed about an arbitrary point, this formula *does not, in general, apply*. There are special exceptions to this rule, but we will not treat them here.

As with a translational system, if the rotational inertia of a rigid body is constant, the right-hand side of the above equation can be written as $\frac{d\vec{H}}{dt} = J \dot{\vec{\omega}}$. In making this simplification, *and considering only motion in one plane,* we can write Newton's second law for a rigid body rotating about its center of mass as:

$$\sum \vec{T} = J \vec{\dot{\omega}}$$

You will often see this equation written as a scalar equation for objects that rotate only in one plane (such as a wheel functioning normally. If you imagine a wheel flying off of a car, you can see that it might have angular velocity about more than one axis of rotation). This scalar version of Newton's second law comes from the fact that if a body rotates only about one axis, its angular velocity (and thus, the torques that produce it) must all be going either into or out of the plane.

In summary, summing the torques about an object's center of mass to see how its *rotational velocity* changes over time is analagous to summing the *forces* applied to an object to see how its *velocity* will change over time.

## Model Construction continued: Free Body Diagrams and Newton's Second Law

Returning to the truck problem, we see that in considering the free-body diagram of the axle, we had not drawn the *torque* from the bearing on the axle. We also did not draw the torque from the bearing on our truck body!! Now that we have introduced the idea of torque and how it extends what we can do with Newton's second law in section 2, two revised free-body diagrams are presented below.

Looking first at the car body: with the addition of "torque" and Newton's second law for angular momentum as tools, you may notice that the car body's gravitational force, which acts at its center of mass, does *not* produce any torque on the car body, but that *all of the other forces we drew will*, since they are located away from the car's center of mass. We also see that the bearings themselves are producing torques on the front and rear axles. So do we need to do a sum of the torques equation for the car body?

We cetainly could! We could also do a sum of the forces in the $y$-direction for the car body. But let's hold off unless (or until) we need to to hunt down unknown parameters or terms, since we observed that the car body does not rotate! The more pressing issue for the car body is still determining the bearing forces $F_{bf}$ and $F_{br}$ that act to slow the car down. **Remember that finding these two forces will give us everything we need to build our final model!!**

For the axle we drew, we can skip (for now) the sum of the forces in the y-direction, because the wheel is not accelerating in that direction, and our output ($v$) doesn't appear to be relevant to that direction. Again, if we need to do that portion of the Newton's second law analysis to eliminate unknowns later, we can. 

For now, looking at the axle requires a sum of the forces in the $x$-direction, which is the same as before. The center of mass of the axle is traveling at the same velocity $v$ as the car. We will label the mass of the axle (including its two wheels) as $m_w$.

$$\sum F_x = m_w\dot{v} = F_{bwf}-F_{ff}$$

The frictional force and the reaction force are both unknown! However, we now have a new Netwon's second law equation that tells us about the *torques* and the *angular velocity* of the wheel. Summing the torques acting on the axle about its center of gravity, this is given by:

$$\sum T \hat{k} = J \dot{\omega} \hat{k} = T_{bwf}\hat{k} + \vec{r}_{wg}\times\vec{F}_{ff}$$

In the above equation, $\vec{r}_{wg}$ represents the position vector from the center of the wheel to the ground, and $\omega$ represents the angular velocity of the wheel and axle. Now, knowing that the friction force vector $F_{ff}$ points directly perpendicular to the position vector connecting the wheel's axis of rotation (also its CG, if we flatten it into a 2D object) and the ground, we can easily calculate the torque of friction on the wheel using the right-hand rule as:

$$\vec{r}_{wg}\times\vec{F}_{ff} =- F_{ff} r \hat{k}$$

This torque is negative because it is in the clockwise direction as drawn (see the discussion of cross products and the right-hand rule above).  Because all of the torques are in the $\hat{k}$ direction, we can write the sum of the torques as a *scalar* equation:

$$T = J \dot{\omega} = T_{bwf} - F_{ff} r$$

You may wonder how this helps us, since $\dot{\omega}$, $T_{bwf}$, and $F_{ff}$ are all unknown! To find out, let's look at the final *element* in our system: the bearing. To analyze the bearing, we must consider Newton's 3rd law of equal and opposite reactions. Any forces we drew on the axle that we said were due to the bearing must act in an equal and opposite sense on the bearing itself. The same can be said for all of the forces from the bearing acting on the car body, and for *torques* as well as forces.


Here, we have finally arrived at a need to *model* the bearing, since we can intuit that the relative motion of its inner "race" (inner portion) and its outer "race" is what causes the friction that slows down the whole car. Our first modeling step is going to be to assume that relative to the mass of the axle and the mass of the car itself, that the *bearing's mass* is insignificant. This means that in writing Newton's second law for the bearing, we can actually write $\sum \vec{F} = 0$, since we are assuming that the bearing's mass is "essentially zero." This means, upon breaking the equation up into x and y components, that:

\begin{align}
F_{bwf} & = F_{bcf} \\
N_{bwf} & = N_{bcf}
\end{align}

In a similar fashion, we can say that the *moment of inertia* of the bearing is insignificant when compared to that of either the car or the axle, and summing torques with this assumption yields:

$$ T_{bwf} = T_{bcf}$$

This is helpful, because it will allow us to connect the Newton's second law equation for the car with the one for the (each) axle. However, it doesn't reveal any new physics. The "new physics" associated with the bearing are that for a lubricated contact, force (or torque, depending on whether the contact is rotational or translational) is proportional to angular (or linear) velocity. Elements such as bearings or lubricated sliding contacts that have *insignificant inertia* and follow this sort of proportional relationship between velocity and force are often called *idealized dampers*. We break from the example once again to learn about idealized dampers!

## Idealized Rotational and Translational Dampers: Force and Torque

Idealized rotational and translational dampers are elements that have insignificant translational and rotational inertia, but that produce a force or torque that resists motion proportionally to the velocity difference between each part of the damper. "True" idealized dampers **do not exist**, because nothing is massless, and almost nothing has a purely linear relationship between force and velocity (or torque and angular velocity). However, *empirical observations* have led scientists and engineers to model many "real" system elements as if they were idealized dampers.

For example, rotational bearings and lubricated sliding contacts are often modeled as idealized dampers, as are translational "shock absorbers" such as those found on a car's suspension. Dampers are often indicated on a drawing using a shorthand sketch of their free body diagram:

Because idealized rotational and translational dampers do not have significant mass, the torque or force (respectively) on each end of the damper must be equal and opposite. 

The magnitude and of the torque on a rotational damper, as drawn above, is given by:

$$ T_b = b(\omega_1 - \omega_2)$$

Where $b$ is the damper's "damping coefficient." This coefficient is often experimentally determined, and its units are $\frac{N\cdot m \cdot s}{rad}$ in SI units so that multiplying the coefficient with an angular velocity in $\frac{rad}{s}$ yields a torque in $N\cdot m$. Note that using the right-hand rule, we can confirm that if $\omega_1$ is larger than $\omega_2$, the torque will point "outwards" from each end of the damper.

Similarly, for a translational damper, the equation for the force produced by the damper is given by:

$$ F_b = b(v_1-v_2)$$

The damping coefficient $b$ for a translational damper has units of $\frac{N\cdot s}{m}$ such that multiplying the coefficient by a velocity in $\frac{m}{s}$ produces a force in $N$. 


## Challenge: Model Construction Continued

While we know that the "real" damper is **not** an idealized damper, it's the best tool we have. It will definitely capture more of the physics than the "dry friction" model we attempted to use at first. If we discover, **during a model evaluation phase**, that this simplification is inappropriate, we can revisit it and add complexity if required.

The decision to model the bearing as an "idealized damper" means we have a way to continue with model construction.  So we will revisit the free-body diagram for the axle. Our Newton's second law equation for the torques on the axle can be rewritten by substituting the "idealized damper" equation in for the torque of the bearing on the axle. One "side" of the bearing is rotating at the speed of the front wheels, $\omega$, and the other side is rotating at the speed of the car's body, or $0 \frac{rad}{s}$. Adapting the "standard" FBD for an idealized damper to our scenario, we can draw:

Looking at the damper (bearing) from the left end, we can ask: "is this drawing consistent with the setup of my system model?" In the FBD of our bearing on the right, we see that we drew the torque from the axle acting on the bearing's *inner race*, which rotates with the axle, as clockwise, or negative. Does this mean our two representations of the bearing are inconsistent? Well, not exactly. We drew our FBDs for the system with torque and force directions defined by what happens when the car is *moving with a positive velocity $v$*. If the car moves at a positive velocity $v$, the wheels, and thus the axle and the inner race of the bearings, will rotate *clockwise*, which represents a negative angular velocity. 

If we substitute a negative angular velocity into our "idealized damper equation," we will see that the torque $T_{bwf}$ flips signs, and takes the direction we drew on our FBD of the bearing (pictured right). So *as drawn*, the two free body diagrams are consistent. Checking this type of consistency is a tricky but necessary step!

Now, to actually *use* the damper equation, we note first that because we model the bearing as an ideal damper, the magnitudes $T_{bwf}=T_{bcf} = T_b$. We can then write:

$$T_b = b\omega_1 = b\omega $$

Where $\omega$ is the angular velocity of the wheel. But what *is the value of* $\omega$ as the car moves? We can find it if we remember that the wheels of the tuck *roll without slip.* Therefore, one revolution of the wheels means that the car has traveled a linear distance equal to the tires' circumference. In fact, if we imagine that the wheel of the car has a string wrapped around it, and we pin that string to the ground and begin to "unroll" the string by allowing the tire to roll along the ground, the amount of string that has unrolled is equivalent to the distance the tire has traveled. Using the relationship between this linear distance and how many radians the tire rotates through can tell us about the relationship between angular and linear velocity.

For our setup, where $x$ points right, $y$ points upwards, and counterclockwise rotation is positive, the distance traveled in the $x$-direction is equal to the radius of the tire multiplied by the angle that the tire has rotated *clockwise*, or in the *negative direction*, under the roll-without-slip condition. As an equation, this can be written:

$$x = -r_{wg}\theta$$.

If we take the time-derivative of this equation, assuming the tire radius $r_{wg}$ stays the same, we can see that:

$$ \dot{x} = v = -r_{wg}\dot{\theta} = -r_{wg}\omega$$

Using this substitution, we can replace any appearance of $\omega_1$, which is *not* one of our desired outputs, with $\frac{-v}{r_{wg}}$. In particular, we can re-write our sum of the torques equation for the *axle* in scalar form (because all $\hat{k}$ cancel), assuming clockwise torques are positive in magnitude. 

$$\sum T = J\dot{\omega} = T_{bwf} + \vec{r}_{wg}{F}_{ff} =  -\frac{b}{r_{wg}}v - \vec{r}_{wg}{F}_{ff} $$

We can further simplify this by realizing that we can substitute $-\frac{\dot{v}}{r_{wg}}$ in for $\dot{\omega}$:

$$\sum T = -J\frac{\dot{v}}{r_{wg}} =  -\frac{b}{r_{wg}}v - \vec{r}_{wg}{F}_{ff} $$

This is a big deal because it allows us to solve for our **unknown friction force** $F_{ff}$, which will allow us to *eliminate it* if we substitute our expression in for another one of our Newton's second law equations. In the sum-of-torques equation for the axle, we can write:

$$F_{ff} = \frac{1}{r_{wg}} \left[ J\frac{\dot{v}}{r_{wg}} +\frac{b}{r_{wg}} v \right] $$


Now, it's time to ask: "does this equation make sense?" This question can be answered in two parts: Do all of the terms in the equation have the same units? Do the signs of the terms make sense?

The answer to the first question is "yes." To answer the second, revisit the free body diagram for the axle, which I have reproduced below.

If the truck were to accelerate quickly in the *positive* x-direction from rest, we would imagine that the magnitude of $F_{ff}$, which is already drawn in a negative direction, would increase. This makes sense. Likewise, if the truck was being held at a constant positive speed $v$, and that speed was large, it also makes sense that the magnitude of $F_{ff}$ would be large and consistent with the *direction* drawn on our diagram. If we were to move the truck *backwards*, we would imagine that the vector $F_{ff}$ would reverse its direction, corresponding with a negative magnitude in our equation.

Now, we can use our expression for the "mysterious" unknown force $F_{ff}$ and substitute it into the *translational* (sum of forces) Newton's second law equation for the axle. Doing this yields the following:

$$\sum F_x = m_w\dot{v} = F_{bwf}-F_{ff} = F_{bwf}-  \frac{1}{r_{wg}} \left[ J\frac{\dot{v}}{r_{wg}} +\frac{b}{r_{wg}} v \right] $$

Cleaned up, we can write this equation as:

$$m_w\dot{v} = F_{bwf}-  \frac{1}{r_{wg}} \left[ \frac{J}{r_{wg}}\dot{v} +\frac{b}{r_{wg}} v \right] $$

We are getting closer to having an equation that *only* has our output $v$ in it, along with known (or measurable) parameters, and you may now recognize what we have as a **differential equation** of the type we determined **experimentally** for the truck in Reading 6. This is great news!

However, there's still the matter of $F_{bwf}$, which is unknown... or is it? We know, based on our FBD of the bearing, that $F_{bwf} = F_{bcf}$ in magnitude, so we may be able to eliminate this unknown term by substituting it into our Newton's second law equation for the truck. Solving the axle's sum of forces equation for $F_{bwf}$ yields:

$$ F_{bwf} = m_w\dot{v} +  \frac{1}{r_{wg}} \left[ \frac{J}{r_{wg}}\dot{v} +\frac{b}{r_{wg}} v \right] $$

Substituting this into our sum of forces equation for the truck's body, which has a mass $m_c$, yields:

$$\sum F_x = m_c \dot{v} = -F_{bcf}-F_{bcr}$$

Using the equivalence of $F_{bcf}=F_{bwf}$, and substituting in our expression, we can write:

$$\sum F_x = m_c \dot{v} = - m_w\dot{v} -  \frac{1}{r_{wg}} \left[ \frac{J}{r_{wg}}\dot{v} +\frac{b}{r_{wg}} v \right]-F_{bcr}$$

This equation has no unknown terms in it, except for $F_{bcr}$. If we assume that the two axles weigh the same, and that the rear wheels also roll without slip, we can say that the two axles have identical dynamics, meaning that $F_{bcr} = F_{bcf}$. Now, we can make a further substitution, and we see that:

$$m_c \dot{v} = - 2m_w\dot{v} -  \frac{2}{r_{wg}} \left[ \frac{J}{r_{wg}}\dot{v} +\frac{b}{r_{wg}} v \right]$$

This is a complete model! Gathering terms, solving for $\dot{v}$, and distributing $r_{wg}$ gives us a nice, clean differential equation:

$$ \dot{v} = -  \frac{1}{m_c+2m_w+2\frac{J}{r_{wg}^2}} \left[  2\frac{b}{r_{wg}^2} v \right]$$

If $b$, $J$, and $r_{wg}$ are all positive (as they should be), this equation has the exact behavior we're looking for! It matches the form of the equation we built in Reading 6, but it allows us to **adjust** physical parameters in the model so that it can *generalize* to different configurations of the truck. For instance, if we wanted to see how the truck would behave if we loaded the bed, we could simply adjust the mass of the truck body, $m_c$, and run a new simulation using Euler integration. **This** is the power of a physical model.

You may notice that our list of independent variables changed a bit since our "first guess" in section 1! The quantities we'd need to know to solve the *differential equation* relating acceleration to velocity are, as a final list:

* truck body mass $m_c$
* wheel/axle mass $m_w$
* bearing damping coefficient $b$
* wheel/axle moment of inertia $J$
* wheel radius $r_{wg}$

This evolution is common. As we work the job of model construction, we often uncover things that are important that we didn't think of at first! It's also true that some quantities we thought *would* be important are not-- notice the lack of the gravitational constant $g$ in this equation!

It may seem like our journey to this differential equation was convoluted, and you may wonder how you could use this same procedure for a different problem. However, the process we followed was actually very simple. The steps in our overall procedure are codified below. This is the same procedure we used in week 1, but modified to include torques as part of our process!

##  Disciplined Process: Using newton's laws to obtain a mechanical system model

To construct a differential equation to serve as a model for a system comprised of "idealized elements" such as masses and dampers, you must:

1. Draw the FBD for each individual element in the system
2. Write Newton's second law for each individual element in the system, using sums of forces and/or torques as appropriate given the motion of your system
3. Eliminate all motion and force terms other than terms multiplied by the system's dependent variables (and their derivatives, as appropriate) using algebraic manipulation. Solving an equation for an unwanted term and substituting into another equation is a good way to eliminate the unwanted term. 
4. Check to make sure that all terms multiplied by your model's dependent variables contain only known or measurable quantities (independent variables).
